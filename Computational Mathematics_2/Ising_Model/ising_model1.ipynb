```python
import numpy as np 
import matplotlib.pyplot as plt 
# plt.style.use(['science', 'notebook', 'grid'])  # disabled because 'science' style is not installed
import numba
from numba import jit, njit, prange
from scipy.ndimage import convolve, generate_binary_structure
```

```python
N = 4
init_random = np.random.random((N, N))
lattice_n = np.zeros((N,N))
lattice_n[init_random >= 0.25] = 1
lattice_n[init_random < 0.25] = -1
plt.imshow(lattice_n)
```

```python
def get_energy(lattice):
    kern = generate_binary_structure(2, 1)
    print(kern)
    kern[1][1] = False
    print(kern)
    arr = -lattice * convolve(lattice, kern, mode='constant')
    return arr.sum()
```

```python
energy = get_energy(lattice_n)
energy
```
